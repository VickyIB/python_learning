```python
# Variables and Types

# Assigning values to Variables
site_name = 'programiz.pro'
print(site_name)

# Changing the Value of a Variable
site_name = 'apple.com'
print(site_name)

print()
print('#' * 20)
print()

# Assigning multiple values to multiple variables
a, b, c = 5, 3.2, 'Hello'
print (a)
print (b)
print (c)

print()
print('#' * 20)
print()

# Converting integer to float
integer_number = 123
float_number = 1.23
new_number = integer_number + float_number
print("Value:",new_number)
print("Data Type:",type(new_number))

print()
print('#' * 20)
print()

# Addition of string and integer Using Explicit Conversion
num_string = '12'
num_integer = 23
print("Data type of num_string before Type Casting:",type(num_string))

# explicit type conversion
num_string = int(num_string)
print("Data type of num_string after Type Casting:",type(num_string))

num_sum = num_integer + num_string
print("Sum:",num_sum)
print("Data type of num_sum:",type(num_sum))
```

```python
# Control flow

# if elif else Statement
number = 0
if number > 0:
    print('Positive number')
elif number <0:
    print('Negative number')
else:
    print('Zero')
print('This statement is always executed')

print()
print('#' * 20)
print()

# for Loop
languages = ['PHP', 'Python', 'React']
for i in languages:
    print(i)
    
print()

language = 'Python'
for x in language:
    print(x)
    
print()

values = range(4)
for i in range(4):
    print(i)
    
print()

# while Loop
number = 1
while number <= 3:
    print(number)
    number = number + 1

print()
print('#' * 20)
print()
 
# break Statement
for i in range(5):
    if i == 3:
        break
    print(i)
    
print()
print('#' * 20)
print()

#continue Statement
for i in range(5):
    if i == 3:
        continue
    print(i)
```

```python
# Strings
name = "Python"
print(name)

# access 1st index element
print(name[1])

# access 4th last element
print(name[-4])

# access character from 1st index to 3rd index
print(name[1:4])

print()

# multiline string 
message = """
Never gonna give you up
Never gonna let you down
"""
print(message)

print()

# Compare Two Strings
str1 = "Hello, world!"
str2 = "I love Swift."
str3 = "Hello, world!"

# compare str1 and str2
print(str1 == str2)
# compare str1 and str3
print(str1 == str3)

print()

# Join Two or More Strings
greet = "Hello, "
name = "Jack"
# using + operator
result = greet + name
print(result)
```

```python
# List
ages = [19, 26, 29]
print(ages)

# access the first element
print(ages[0])   # Python

# access the third element
print(ages[2])

print()

# Add Elements to a Python List
print('Original List:', ages)

# using append method 
ages.append(40)

print('Updated List:', ages)

# changing the third item to 'Blue'
ages[2] = 10

print('Updated List:', ages)

# remove 4 from the list
ages.remove(26)
print('Updated List:', ages)
```

```python
# Dictionary
country_capitals = {
  "Germany": "Berlin", 
  "Canada": "Ottawa", 
  "England": "London"
}

print(country_capitals)

print()

# access the value of keys
print(country_capitals["Germany"])
print(country_capitals["England"])

print()

# add an item with "Italy" as key and "Rome" as its value
country_capitals["Italy"] = "Naples"
print(country_capitals)

print()

# change the value of "Italy" key to "Rome"
country_capitals["Italy"] = "Rome"
print(country_capitals)

print()

# delete item having "Germany" key
del country_capitals["Germany"]
print(country_capitals)

print()

# print dictionary values one by one
for country in country_capitals:
    capital = country_capitals[country]
    print(capital)

print()

# print dictionary keys one by one
for country in country_capitals:
    print(country)

print()

# print dictionary values one by one
for country in country_capitals:
    capital = country_capitals[country]
    print(capital)
    
print()

# clear the dictionary
country_capitals.clear()
print(country_capitals) # print dictionary keys one by one
for country in country_capitals:
    print(country)
```

```python
# Operators
# Arithmetic Operators
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

```python
# Comparison Operators
a = 10
b = 3

print("Equal:", a == b)
print("Not Equal:", a != b)
print("Greater than:", a > b)
print("Less than:", a < b)
print("Greater than or equal to:", a >= b)
print("Less than or equal to:", a <= b)
```
